### Multiplicity (track, bounds, length, and matched; mu, pi, p, g, e) Truth

- In this cell, we are counting multiplicity of truth (truth_particles) muons, pions, protons, photons, and electrons.
- `semantic_type` is 1 (which are tracks).
- `is_contained` condition is set to 1.
- `length` should be greater than `5.0` cm.
- `matched` condition is set to 1.
- Apply bounds using `start_point` and `end_point` variables from the `Particle` dataset.

```python
import sys
import yaml
import h5py
import numpy as np
import matplotlib.pyplot as plt

# Define epsilon for the distance from the wall (in cm)
epsilon = 5.0

# Define bounds (in cm) with epsilon applied
Xnegbound, Xposbound = -63.931 + epsilon, +63.931 - epsilon
Ynegbound, Yposbound = -329.076 + epsilon, -204.924 - epsilon
Znegbound, Zposbound = +1235.462 + epsilon, +1364.538 - epsilon

# Create an empty list to store counts for Particle
selected_particle_counts = []

# Loop over the HDF5 files
for file_num in range(1, 2):  # Files from 1 to 2
    # Define the data keys for the HDF5 file
    #data_keys = f'/app/data/output_ana/output_27023276_{file_num}-larcv_mlreco_ana.h5'
    data_keys = f'output_27024464_{file_num}-larcv_mlreco_ana.h5'

    
    # Open the HDF5 file to retrieve the total number of entries
    with h5py.File(data_keys, 'r') as h5_file:
        total_entries = len(h5_file['events'])

    # Define a list of entry numbers to load
    entry_numbers = list(range(total_entries))

    # Load and process each entry
    for ENTRY_NUM in entry_numbers:
        # Define the analysis configuration
        cfg = f'''
        analysis:
          iteration: -1
          log_dir: log_trash
          data_builders:
            - ParticleBuilder
        reader:
          name: HDF5Reader
          file_keys: DATA_KEY
        '''.replace('DATA_KEY', data_keys)

        cfg = yaml.safe_load(cfg)

        # Initialize the AnaTools manager
        manager = AnaToolsManager(cfg)
        manager.initialize()

        # Load the specific entry
        data, result = manager.forward(iteration=ENTRY_NUM)
        manager.load_representations(data, result, mode='all')

        # Retrieve reconstructed particles for this entry
        particles = result['truth_particles'][0]

        # Count the number of particles with specific PDG codes, bounds on start_point and end_point,
        # semantic_type, and is_contained
        selected_pdg_codes = [13, 22, 11, 2212, 211]
        particle_count = sum(1 for particle in particles if particle.pdg_code in selected_pdg_codes
                             and Xnegbound <= particle.start_point[0] <= Xposbound
                             and Ynegbound <= particle.start_point[1] <= Yposbound
                             and Znegbound <= particle.start_point[2] <= Zposbound
                             and particle.semantic_type == 1
                             and particle.length > 5.0
                             and particle.matched == 1)

        # Append the counts to the list
        selected_particle_counts.append(particle_count)

        # Print the information
#         for particle in particles:
#             if particle.pdg_code in selected_pdg_codes:
#                 pdg = particle.pdg_code
#                 is_contained = particle.is_contained
#                 start_point = particle.start_point
#                 end_point = particle.end_point
#                 print(f'id (entry): {ENTRY_NUM}, count: {int(particle.pdg_code in selected_pdg_codes)}, pdg: {pdg}, is_contained: {is_contained}, ' \
#                       f'start_point: [{start_point[0]:.3f}, {start_point[1]:.3f}, {start_point[2]:.3f}], ' \
#                       f'end_point: [{end_point[0]:.3f}, {end_point[1]:.3f}, {end_point[2]:.3f}]')
                


# Calculate the total number of selected particles
total_selected_particles = sum(selected_particle_counts)

# Create a histogram of Particle counts with red color
bins = np.arange(1, 11, 1)  # Start from 1, end at 11 (exclusive) to include 10, with step 1
plt.hist(selected_particle_counts, bins=bins, edgecolor='black', alpha=1.0, color='red', label='Truth')

plt.xlabel('Particle Count')
plt.ylabel('Frequency')
plt.title('Histogram of mu, pi, p, e, g (tracks, bounds and matched)')

# Set x-axis ticks at integer positions
plt.xticks(np.arange(1, 11, 1))

# Add a legend
plt.legend()

# Calculate and display statistics for Selected Particles
mean_count_selected_particle = np.mean(selected_particle_counts)
std_dev_selected_particle = np.std(selected_particle_counts)
stats_text_selected_particle = f'Selected Particles\nEntries: {total_selected_particles}\nMean: {mean_count_selected_particle:.2f}\nStd Dev: {std_dev_selected_particle:.2f}'

# Create a statbox and place it on the left side of the canvas
plt.text(0.645, 0.60, stats_text_selected_particle, transform=plt.gca().transAxes, bbox=dict(facecolor='white', alpha=0.5), ha='left', va='center')

# Save the plot as selected_particles_histogram.png in the current directory
plt.savefig('mult_mu_pi_p_g_e_track_bounds_length_matched.png')
plt.show()
```
